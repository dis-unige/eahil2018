## Extrating the useful metadata from PubMed XML

PubMed data is freely accessible from the NLM FTP following the information available here: https://www.nlm.nih.gov/databases/download/pubmed_medline.html

The whole database is downloadable in form of hundreds of compressed XML files (medline17n[0001-0892].xml.gz). At the moment of our study (27.10.2017) we obtained 892 gz files (20Gb in total)

The elements of XML files are described here : https://www.nlm.nih.gov/bsd/licensee/elements_descriptions.html

We try some methods to parse the files with this Jupyter Notebook but we abandon because we run intop memeroy limits. Finaly we parse the XML files and stract only useful informations via a Cygwin shell script using the XMLstarlet toolkit (https://en.wikipedia.org/wiki/XMLStarlet) :

```bash
#!/bin/bash
for file in xml/*.gz
do
	xmlstarlet sel -t -m /PubmedArticleSet/PubmedArticle -v "concat(MedlineCitation/PMID,'|',MedlineCitation/Article/Journal/ISSN,'|',MedlineCitation/MedlineJournalInfo/ISSNLinking,'|',MedlineCitation/Article/Journal/ISOAbbreviation,'|',MedlineCitation/Article/Journal/JournalIssue/PubDate/Year,'|',MedlineCitation/Article/Journal/JournalIssue/PubDate/MedlineDate,'|',MedlineCitation/Article/Journal/JournalIssue/Volume,'|',MedlineCitation/Article/Journal/JournalIssue/Issue,'|',MedlineCitation/Article/Pagination/MedlinePgn,'|',PubmedData/ArticleIdList/ArticleId[@IdType='doi'],'|',PubmedData/ArticleIdList/ELocationID[@EIdType='doi'],'|',PubmedData/PublicationStatus,'|',MedlineCitation/@Status)" -n "$file" > "csv/$file.csv"
done
```

Then we concat all the CSV results in one file via the unix tool "cat" :
```
cat *.csv > pubmed_extraction_all.csv
```
The file has 1Gb and is ready to be used on others
