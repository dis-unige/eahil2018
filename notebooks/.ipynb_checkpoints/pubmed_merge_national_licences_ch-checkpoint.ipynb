# Swiss National Licences

We test the coverage of PubMed by the national licences project http://consortium.ch/nationallizenzen/?lang=en :

"National licences improve the nationwide availability of e-resources in Switzerland. The project concentrates on acquiring complete and frequently used journal archives. Eligible users include researchers, students and university staff members, the ETH Domain, cantonal libraries and other non-commercial institutions of higher research and education. Private users with a permanent residence in Switzerland, have also access since the end of 2016."

### Content available

"In order to guarantee a relevant and varied portfolio for the national licences, established academic publishers were selected for the project. These national licences form a journal portfolio that is used comprehensively and frequently. Access until the recent past (2015) is possible."

 - Springer Journals Archive
     - number of articles: 3,600,000
     - years covered: 1832-2012
     - moving wall embargo: 4 years
 - Oxford Journals Archive
     - number of articles: 800,000
     - years covered: 1895-2015
     - moving wall embargo: 3 years
 - De Gruyter Journals Archive
     - number of articles: 483,000
     - years covered: 1826-2015
     - moving wall embargo: 2 years
 - Cambridge Journals Archive
     - number of articles: 1,123,000
     - years covered: 1770-2015
     - moving wall embargo: 5 years

Titles lists of those 4 publishers were downloaded on 02.05.2018 form this URL into the source folder "national_licences_ch": http://consortium.ch/title-lists-for-swiss-national-licences/?lang=en

### National licences outside this project also taken into account:

## Merge PubMed and National licences CH journals information

```python
# Open Pubmed data
import pandas as pd
pubmed = pd.read_csv('data/sources/pubmed/pmid_issn_issnl_year_clean.csv.gz', delimiter='\t',
                     dtype={'pmid': 'int', 'issn': 'object', 'issnl': 'object', 'year': 'int'}, 
                     header=0)
pubmed
```

```python
# import doaj data
licences_ch = pd.read_csv('data/sources/national_licences_ch/national_licences_ch_ready_to_pubmed_merge.csv', delimiter='\t', header=0,
                       dtype={'issnl': 'object', 'syear': 'int', 'eyear': 'int'},
                       usecols=('issnl', 'syear', 'eyear'))
licences_ch
```

```python
# normalize ISSNL
# remove blanks
licences_ch['issnl'] = licences_ch['issnl'].str.strip()
# convert to lower case
licences_ch['issnl'] = licences_ch['issnl'].str.lower()
# rename years
licences_ch = licences_ch.rename(columns = {'syear': 'lch_syear'})
```

```python
# merge PubMed with journals data by ISSNL
pubmed = pubmed.merge(licences_ch, on = 'issnl', how='left')
pubmed
```

```python
pubmed = pubmed.rename(columns = {'eyear': 'lch_eyear'})
pubmed
```

```python
# Add LCH column
pubmed['LCH'] = 0
# Calculate if years are in range
pubmed.loc[(pubmed['year'] >= pubmed['lch_syear']) & (pubmed['year'] <= pubmed['lch_eyear']), 'LCH'] = 1
pubmed
```

```python
# sort values and drop duplicates to keep the good rows (with 1)
pubmed = pubmed.sort_values(by=['LCH', 'pmid'], ascending=[False, True])
pubmed
```

```python
# drop duplicates
pubmed = pubmed.drop_duplicates(subset='pmid')
pubmed
```

```python
# PMIDs covered by LCH journals 
pubmed.loc[pubmed['LCH'] == 1].shape
```

```python
# export the result
pubmed.to_csv('data/results/pubmed_lch.csv.gz', sep='\t', encoding='utf-8', index=False, compression='gzip')
pubmed[['pmid', 'LCH']].to_csv('data/results/pubmed_lch_short.csv.gz', sep='\t', encoding='utf-8', index=False, compression='gzip')
```

```python
# export the result for positives
# pubmed.loc[pubmed['LCH'] == 1].to_csv('data/results/pubmed_lch_positives.csv.gz', sep='\t', encoding='utf-8', index=False, compression='gzip')
pubmed.loc[pubmed['LCH'] == 1][['pmid', 'LCH']].to_csv('data/results/pubmed_lch_positives_short.csv.gz', sep='\t', encoding='utf-8', index=False, compression='gzip')
```
