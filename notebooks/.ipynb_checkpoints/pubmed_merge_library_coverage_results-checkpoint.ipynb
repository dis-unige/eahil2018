## Results of library coverage 
### PubMed articles covered by print and e-journals of Geneva University Library collection

PubMed metadata was merged with print journals information from catalog and e-journals information from link resolver (including free journals). The key was the ISSNL and the year of publication. This merge it could be more precise including volume and issue information but this metadata are not well coded and we prefer to stay at the year level. We guess that the false positives and false negatives will be ballanced.

```python
# Open PubMed file
import pandas as pd
pubmed = pd.read_csv('data/results/pubmed_pj_ej.csv.gz', delimiter='\t',
                     dtype={'pmid': 'int', 'year': 'int', 'PJ': 'int', 'EJ': 'int', 'is_free': 'float'}, header=0)
pubmed
```

```python
pubmed.shape
```

```python
# check NA
pubmed.loc[pubmed['pmid'].isnull()]
```

```python
# check NA
pubmed.loc[pubmed['year'].isnull()]
```

```python
# check NA
pubmed.loc[pubmed['PJ'].isnull()]
```

```python
# check NA
pubmed.loc[pubmed['EJ'].isnull()]
```

```python
# check NA
pubmed.loc[pubmed['is_free'].isnull()]
```

```python
# Replace NaN by 0 to simplify
pubmed = pubmed.fillna(0)
pubmed['is_free'] = pubmed['is_free'].astype(int)
pubmed
```

## Counting PMIDs

```python
# PMIDs covered by print journals
pm = pubmed['pmid'].shape
pj = pubmed.loc[pubmed['PJ'] == 1]['pmid'].shape
print (str(pj[0]) + ' / ' + str(pm[0]) + ' = ' + str(pj[0]/pm[0]))
```

```python
# PMIDs covered by e-journals
ej = pubmed.loc[pubmed['EJ'] == 1]['pmid'].shape
print (str(ej[0]) + ' / ' + str(pm[0]) + ' = ' + str(ej[0]/pm[0]))
```

```python
# PMIDs covered by print or e-journals
epj = pubmed.loc[(pubmed['PJ'] == 1) | (pubmed['EJ'] == 1)]['pmid'].shape
print (str(epj[0]) + ' / ' + str(pm[0]) + ' = ' + str(epj[0]/pm[0]))
```

```python
# PMIDs not covered by print or e-journals
nepj = pubmed.loc[(pubmed['PJ'] == 0) & (pubmed['EJ'] == 0)]['pmid'].shape
print (str(nepj[0]) + ' / ' + str(pm[0]) + ' = ' + str(nepj[0]/pm[0]))
```

```python
# PMIDs covered by free e-journals
fej = pubmed.loc[(pubmed['EJ'] == 1) & (pubmed['is_free'] == 1)]['pmid'].shape
print (str(fej[0]) + ' / ' + str(pm[0]) + ' = ' + str(fej[0]/pm[0]))
```

```python
# PMIDs covered by non free e-journals
nfej = pubmed.loc[(pubmed['EJ'] == 1) & (pubmed['is_free'] == 0)]['pmid'].shape
print (str(nfej[0]) + ' / ' + str(pm[0]) + ' = ' + str(nfej[0]/pm[0]))
```

```python
# PMIDs covered by print journals only (not e-journals)
pjo = pubmed.loc[(pubmed['EJ'] == 0) & (pubmed['PJ'] == 1)]
pjo.shape
```

```python
# PMIDs covered by e-journals only (not print journals)
ejo = pubmed.loc[(pubmed['EJ'] == 1) & (pubmed['PJ'] == 0)]
ejo.shape
```

```python
# PMIDs covered by e-journals and print journals)
eapj = pubmed.loc[(pubmed['EJ'] == 1) & (pubmed['PJ'] == 1)]
eapj.shape
```

```python
# PMIDs covered by free e-journals only (not e-journals payed)
fejo = pubmed.loc[(pubmed['EJ'] == 1) & (pubmed['is_free'] == 1)]
fejo.shape
```

```python
# PMIDs covered by non free e-journals only (not free e-journals)
nfejo = pubmed.loc[(pubmed['EJ'] == 1) & (pubmed['is_free'] == 0)]
nfejo.shape
```

```python
# PMIDs non covered by print or e-journals
nepj = pubmed.loc[(pubmed['PJ'] == 0) & (pubmed['EJ'] == 0)]
nepj.shape
```

## Merging counts to produce graphics

```python
import pandas as pd
# open files exported with counts
myfilein = 'data/results/pmids_library_pj_only_counts_by_year.csv'
pjo_counts = pd.read_csv(myfilein, sep='\t', names=('Year', 'PMIDs covered only by print journals collection') , header=None, encoding='utf-8')
myfilein = 'data/results/pmids_library_ej_only_counts_by_year.csv'
ejo_counts = pd.read_csv(myfilein, sep='\t', names=('Year', 'PMIDs covered only by e-journals collection') , header=None, encoding='utf-8')
myfilein = 'data/results/pmids_library_ej_and_pj_counts_by_year.csv'
eapj_counts = pd.read_csv(myfilein, sep='\t', names=('Year', 'PMIDs covered by both print and e-journals collections') , header=None, encoding='utf-8')
myfilein = 'data/results/pmids_library_free_ej_only_counts_by_year.csv'
fejo_counts = pd.read_csv(myfilein, sep='\t', names=('Year', 'PMIDs covered by free e-jourals only') , header=None, encoding='utf-8')
myfilein = 'data/results/pmids_library_non_free_ej_only_counts_by_year.csv'
nfejo_counts = pd.read_csv(myfilein, sep='\t', names=('Year', 'PMIDs covered by non free e-jourals only') , header=None, encoding='utf-8')
myfilein = 'data/results/pmids_library_not_pj_not_ej_counts_by_year.csv'
nepj_counts = pd.read_csv(myfilein, sep='\t', names=('Year', 'PMIDs not covered by any journal collection') , header=None, encoding='utf-8')
```

```python
# merge
pmid_counts_all = pjo_counts.merge(eapj_counts, on='Year', how='outer')
pmid_counts_all = pmid_counts_all.merge(ejo_counts, on='Year', how='outer')
pmid_counts_all = pmid_counts_all.merge(nepj_counts, on='Year', how='outer')
pmid_counts_all
```

```python
# exclude 2018
pmid_counts_all = pmid_counts_all.loc[pmid_counts_all['Year'] <= 2017]
```

```python
pmid_counts_all = pmid_counts_all.fillna(0)
```

```python
pmid_counts_all = pmid_counts_all.set_index('Year')
pmid_counts_all = pmid_counts_all.sort_index()
pmid_counts_all
```

```python
# merge but separate free and non free e-journals
pmid_counts_all_fj = pjo_counts.merge(nfejo_counts, on='Year', how='outer')
pmid_counts_all_fj = pmid_counts_all_fj.merge(fejo_counts, on='Year', how='outer')
pmid_counts_all_fj = pmid_counts_all_fj.merge(nepj_counts, on='Year', how='outer')
pmid_counts_all_fj
```

```python
# exclude 2018
pmid_counts_all_fj = pmid_counts_all_fj.loc[pmid_counts_all_fj['Year'] <= 2017]
```

```python
pmid_counts_all_fj = pmid_counts_all_fj.fillna(0)
```

```python
pmid_counts_all_fj = pmid_counts_all_fj.set_index('Year')
pmid_counts_all_fj = pmid_counts_all_fj.sort_index()
pmid_counts_all_fj
```

```python
# plot year counts
import matplotlib.pyplot as plt  
%matplotlib inline
myfileoutfig = 'data/results/pmids_library_coverage_newcolors.png'
pmid_counts_all_plot1 = pmid_counts_all
ax = pmid_counts_all_plot1.plot.area(fontsize=80, figsize=(100,50), legend=True, color=['hotpink', 'pink', 'crimson', 'gray'])
ax = ax.legend(loc=2, prop={'size': 80})
fig = ax.get_figure()
fig.set_size_inches(100, 50)
fig.savefig(myfileoutfig, dpi=50)
```

```python
# limit to 1945 >
pmid_counts_1945_1 = pmid_counts_all.iloc[144:]
pmid_counts_1945_1
```

```python
# plot year counts for 1945 ->
myfileoutfig = 'data/results/pmids_library_coverage_after_1945_newcolors.png'
ax = pmid_counts_1945_1.plot.area(fontsize=80, figsize=(100,50), legend=True, color=['hotpink', 'pink', 'crimson', 'gray'])
ax = ax.legend(loc=2, prop={'size': 80})
fig = ax.get_figure()
fig.set_size_inches(100, 50)
fig.savefig(myfileoutfig, dpi=50)
```

```python
# pie chart 1
myfileoutfig = 'data/results/pmids_library_coverage_all_pie_chart_newcolors.png'
labels = 'print journals only', 'print and e-journals', 'e-journals only', 'PMIDs not covered'
sizes = [pmid_counts_all['PMIDs covered only by print journals collection'].sum(),
         pmid_counts_all['PMIDs covered by both print and e-journals collections'].sum(),
         pmid_counts_all['PMIDs covered only by e-journals collection'].sum(),
         pmid_counts_all['PMIDs not covered by any journal collection'].sum()]
explode = (0.1, 0.1, 0.1, 0.1)  # only "explode" the 2nd slice (i.e. 'Hogs')
colors = ['hotpink', 'pink', 'crimson', 'gray']
fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=0, colors=colors)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

fig1.savefig(myfileoutfig, dpi=100)
plt.show()
```

```python
# pie chart 2
myfileoutfig = 'data/results/pmids_library_coverage_all_free_journals_separated_pie_chart_newcolors.png'
labels = 'print journals only', 'non free e-journals', 'free e-journals', 'PMIDs not covered'
sizes = [pmid_counts_all_fj['PMIDs covered only by print journals collection'].sum(),
         pmid_counts_all_fj['PMIDs covered by non free e-jourals only'].sum(),
         pmid_counts_all_fj['PMIDs covered by free e-jourals only'].sum(),
         pmid_counts_all_fj['PMIDs not covered by any journal collection'].sum()]
explode = (0.1, 0.1, 0.1, 0.1)  # only "explode" the 2nd slice (i.e. 'Hogs')
colors = ['hotpink', 'crimson', 'purple', 'gray']
fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=0, colors=colors)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

fig1.savefig(myfileoutfig, dpi=100)
plt.show()
```
