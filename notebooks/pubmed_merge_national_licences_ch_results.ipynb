```python
# Open Last file after restart Kernel
import pandas as pd
pubmed = pd.read_csv('data/results/pubmed_library_unpaywall_pmc_doaj_lch_pmids_year.csv.gz', delimiter='\t',
                     dtype={'pmid': 'int', 'year': 'int'}, compression='gzip', header=0)
pubmed
```

```python
# PMIDs in LCH
pm = pubmed['pmid'].shape
lch = pubmed.loc[pubmed['LCH'] == 1]['pmid'].shape
print (str(lch[0]) + ' / ' + str(pm[0]) + ' = ' + str(lch[0]/pm[0]))
```

```python
# PMIDs in LCH and OA gold
lch_and_oa_gold = pubmed.loc[(pubmed['LCH'] == 1) & ((pubmed['DOAJ'] == 1) | (pubmed['unpaywall_oa_gold'] == 1))]['pmid'].shape
print (str(lch_and_oa_gold[0]) + ' / ' + str(pm[0]) + ' = ' + str(lch_and_oa_gold[0]/pm[0]))
print (str(lch_and_oa_gold[0]) + ' / ' + str(lch[0]) + ' = ' + str(lch_and_oa_gold[0]/lch[0]))
```

```python
# PMIDs in LCH and OA green
lch_and_oa_green = pubmed.loc[(pubmed['LCH'] == 1) & ((pubmed['PMC'] == 1) | (pubmed['unpaywall_oa_green'] == 1))]['pmid'].shape
print (str(lch_and_oa_green[0]) + ' / ' + str(pm[0]) + ' = ' + str(lch_and_oa_green[0]/pm[0]))
print (str(lch_and_oa_green[0]) + ' / ' + str(lch[0]) + ' = ' + str(lch_and_oa_green[0]/lch[0]))
```

```python
# PMIDs in LCH and OA gold or green
lch_and_oa = pubmed.loc[(pubmed['LCH'] == 1) & ((pubmed['PMC'] == 1) | (pubmed['unpaywall_oa_green'] == 1) | (pubmed['DOAJ'] == 1) | (pubmed['unpaywall_oa_gold'] == 1))]['pmid'].shape
print (str(lch_and_oa[0]) + ' / ' + str(pm[0]) + ' = ' + str(lch_and_oa[0]/pm[0]))
print (str(lch_and_oa[0]) + ' / ' + str(lch[0]) + ' = ' + str(lch_and_oa[0]/lch[0]))
```

```python
# PMIDs in LCH and OA gold and green
lch_and_oa2 = pubmed.loc[(pubmed['LCH'] == 1) & ((pubmed['PMC'] == 1) | (pubmed['unpaywall_oa_green'] == 1)) & ((pubmed['DOAJ'] == 1) | (pubmed['unpaywall_oa_gold'] == 1))]['pmid'].shape
print (str(lch_and_oa2[0]) + ' / ' + str(pm[0]) + ' = ' + str(lch_and_oa2[0]/pm[0]))
print (str(lch_and_oa2[0]) + ' / ' + str(lch[0]) + ' = ' + str(lch_and_oa2[0]/lch[0]))
```

```python
# PMIDs in LCH and not OA
lch_not_oa = pubmed.loc[(pubmed['LCH'] == 1) & (pubmed['PMC'] == 0) & (pubmed['unpaywall_oa_green'] == 0) & (pubmed['DOAJ'] == 0) & (pubmed['unpaywall_oa_gold'] == 0)]['pmid'].shape
print (str(lch_not_oa[0]) + ' / ' + str(pm[0]) + ' = ' + str(lch_not_oa[0]/pm[0]))
print (str(lch_not_oa[0]) + ' / ' + str(lch[0]) + ' = ' + str(lch_not_oa[0]/lch[0]))
```

```python
# PMIDs not OA and not LCH
not_oa_and_not_lch = pubmed.loc[(pubmed['LCH'] == 0) & (pubmed['PMC'] == 0) & (pubmed['unpaywall_oa_green'] == 0) & (pubmed['DOAJ'] == 0) & (pubmed['unpaywall_oa_gold'] == 0)]['pmid'].shape
print (str(not_oa_and_not_lch[0]) + ' / ' + str(pm[0]) + ' = ' + str(not_oa_and_not_lch[0]/pm[0]))
```

## Merging counts to produce graphics

```python
import pandas as pd
# open files exported with counts

# total pmids by year
total_pmid_counts = pd.read_csv('data/results/pmids_by_date_corr.tsv', sep='\t', names=('Year', 'PMIDs') , header=0, encoding='utf-8')

# total LCH
total_lch_counts = pd.read_csv('data/results/pmids_lch_counts_by_year.csv', sep='\t', names=('Year', 'PMIDs LCH') , header=None, encoding='utf-8')

# total OA
total_gold_counts = pd.read_csv('data/results/pmids_total_oa_gold_counts_by_year.csv', sep='\t', names=('Year', 'PMIDs OA Gold') , header=None, encoding='utf-8')
total_green_not_gold_counts = pd.read_csv('data/results/pmids_total_oa_green_not_gold_counts_by_year.csv', sep='\t', names=('Year', 'PMIDs OA Green') , header=None, encoding='utf-8')

# LCH not OA
total_lch_not_oa_counts = pd.read_csv('data/results/pmids_lch_not_oa_counts_by_year.csv', sep='\t', names=('Year', 'PMIDs LCH') , header=None, encoding='utf-8')

# not OA not LCH
total_not_oa_counts = pd.read_csv('data/results/pmids_total_not_oa_not_lch_counts_by_year.csv', sep='\t', names=('Year', 'PMIDs not OA not LCH') , header=None, encoding='utf-8')
```

```python
# merge oa total
pmid_counts_all_tot = total_gold_counts.merge(total_green_not_gold_counts, on='Year', how='outer')
pmid_counts_all_tot = pmid_counts_all_tot.merge(total_lch_not_oa_counts, on='Year', how='outer')
pmid_counts_all_tot = pmid_counts_all_tot.merge(total_not_oa_counts, on='Year', how='outer')
pmid_counts_all_tot
```

```python
pmid_counts_all_tot = pmid_counts_all_tot.fillna(0)
```

```python
pmid_counts_all_tot = pmid_counts_all_tot.set_index('Year')
pmid_counts_all_tot = pmid_counts_all_tot.sort_index()
pmid_counts_all_tot
```
